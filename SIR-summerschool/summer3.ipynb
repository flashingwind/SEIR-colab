### クラス

- 「オブジェクト指向」


- 特別な型(のようなもの)がある: クラスという
- 自分で作れる型のようなもの．変数や関数をまとめて管理する．使うときはインスタンス化して変数に入れる
- 今後使うクラスはnp.ndarrayクラスだけ(ベクトル・行列)
- なので，定義のしかたは忘れてもいいが，一応解説する

下記ようなクラス`Counter`を作ってみる．

- 例としてカウンターを作った
- 加算する関数と，カウンターの値を保持する変数を入れてある
- 同じクラスから作ったインスタンスには同じ変数と関数が入っているが，変数の内容は独立
- インスタンス関数の第1引数は`self`としなければならない．呼び出し時には省略するが，自分自身(インスタンス)が自動的に渡される．
- クラスやインスタンスの中身を見たいときは`.`を使う．

```python
class Counter:
  def __init__(self, name, value=0):#初期設定をする特別なメソッド．self以外の引数は自由．'='でデフォルト値を指定
    self.name=name
    self.value=value

  def add_one(self):
    self.value+=1
    return self.value
```

- 定義だけでは何も起きない
- ふつうはそのまま使うのではなく，実体化して使う: インスタンス化
- `float`型そのものを使うのではなく，`float`型のデータを使うのと同じ(？)
- `__init__()`はインスタンス化時に呼ばれるが，この名前で呼んではいけない
- インスタンスを作ったら，適当な変数に入れてつなぎ止める．

```python
#インスタンス化: クラス名(……)という特殊な関数を使う．実際には__init__()が呼ばれ，インスタンスが返る．ここではRという変数に入れる
c1=Counter('カウンター1', 1)
c2=Counter('カウンター2')#省略すると，デフォルト値が使われる
print("type(c1)=",type(c1))

#インスタンス変数を参照
print("c1のname",c1.name)
print("c1のvalue=",c1.value)

#メソッドを使ってみる．インスタンス変数がどうなるか？
print(c1.add_one())#selfは渡さない
print(c1.add_one())#selfは渡さない
print(c2.add_one())#selfは渡さない
print("c1のvalue=",c1.value)
print("c2のvalue=",c2.value)

#インスタンス変数を直接変更するインチキができる．どうなるか？
c1.value=100
print("c1のvalue=",c1.value)
```
