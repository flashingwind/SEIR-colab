```python
import pandas as pd
import matplotlib.pyplot as plt


df0 = pd.read_csv('drive/MyDrive/Colab Notebooks/data/患者テーブル.csv')
print(df0.keys())

#df = df0.sort_values("発症→確定タイムラグ")
#df = pd.to_numeric(df["発症→確定タイムラグ"], errors="coerce").to_list()
#print(df)
#plt.hist(df,bins=10,range=(0,10))
#plt.show()

df = df0.sort_values('発症_年月日')
df['発症_年月日']=pd.to_datetime(df['発症_年月日'], format='%Y/%m/%d')
df['公表_年月日']=pd.to_datetime(df['公表_年月日'], format='%Y/%m/%d')
#df = pd.to_numeric(df["確定→公表タイムラグ"], errors="coerce").to_list()
print(df['発症_年月日'])
print(df['公表_年月日'])
df['diff']=df['公表_年月日']-df['発症_年月日']
df['diff']=df['diff'].dt.days
print(df['diff'])

plt.hist(df['diff'],bins=10,range=(0,10))
plt.show()
```
