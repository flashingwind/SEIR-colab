### `list`とは

- 順番は保存される
- `[値, ……]`で作る
- `a[index]`で参照できる．`index`は0番目からはじまる
- ネスト: ネストした場合`m[3][0]`のように参照できる
- 等差数列などを専用の関数で生成できる
- `len(……)`で長さ

```python
l=['a','b','c']
m=[[1,2,3,4]]

print(l[0])
n=l+m
print(n)
print("l",l)
#l.append(m)
l.insert(len(l),m)
print("l",l)
print(len(l))


o=[[0],
  [1],
  [2]]
print(o)
print(len(o)-1)
```

## `dict`とは

- 'dictionary'の`dict`
- `{key:value,……}`で作る
- `key`と`value`のペアを保存する
  - `list`は`i`番目→`value`を保存する
  - `list`は自然数しか使えなかった
  - `dict`の場合，あらゆる値を`key`にできる
  - ただし，順番は保存されない……かも
- `a[key]`で参照できる


```python
d={'munenori':'a','akihide':'b'}#特に1文字である必要は無い
print(d['munenori'])

d={'munenori':{'kokugo':'a','sansu':'b'},'akihide':{'kokugo':'a','sansu':'b'}}#valueは複数持てないので，そういうときは`dict`や`list`をvalueにする
print(d['munenori'])
print(d['munenori']['kokugo'])
```

### `for-in`文

- くり返しを書く．通常`range(……)`を使う
- `list`や`dict`など．コレクションに対して処理をする

```python
#ただくり返したいとき————range()
for t in range(10):#もっともシンプルな例では，0=<t<MAX，公差=1となる
    print(t)

import numpy as np
for t in np.arange(0,1,0.1):
    print(t)

#x=np.arange(0,1,0.1)
#plt.plot(x,y)

print()#改行

for t in range(0,10,1):#初期値0 <= t < 最大値10，公差1．いちいち指定した．同じ結果になる
    print(t)
```

```python
#listをスキャンする
l=['a','b','c']

for value in l: #一般的なfor-inでは何番目かわからない
    print(value)


#C
#for(i=0;i<=2; i++){
#for(i=0;i<3; i++){
#    print(f[i]);
#}

for i,value in enumerate(l): #enumerate()を使うと番号をふれる．(key,value)のtupleの列が出てくる
    l[i]=value+"A" #文字列同士は+で結合できる．変更したいときは元のlistに代入

print()

for value in l: #変更されている
    print(value)
```

```python
#dictをスキャンする
d={'A':'a','B':'b','C':'c'}
for key in d: #一般的なfor-inではkeyが出てくる
    print(key+':'+d[key])

print()

for key,value in d.items():
    print(key+':'+value)
```

### スライス

- `list`の後に`[i]`を付けると`i`番目のデータを取り出せる
- `list`に付ける`[……]`の中に特殊な記法で範囲(等差数列)を書ける
- `[初期値:最大値の1つ前]`
- または`[初期値:最大値の1つ前:公差]`

```python
l=['a','b','c']
print(l[0])
print(l[0:2])#range()同様，初期値:最大値+1:公差．
```

```python
d={'cat':{1:'komichi',2:'komugi'},'dog':{1:'sakura',2:'sumire'}}
#for key in d:
print(d['cat'][1])


def simple(d):
  for key,value in d.items():
    print(key)#,':',value)
    for key2,value2 in value.items():
      print(key2,':',value2)

```

```python
d={'cat':{1:'komichi',2:'komugi'},'dog':{1:'sakura',2:'sumire'}}
def remove(d,animal):
  z=d[animal]
  #print('d[animal]',animal,':',d[animal])
  del d[animal]
  #print('削除後',d)
  simple(d)
  return z

remove(d,'cat')
```
