```
%matplotlib inline
import matplotlib.pyplot as plt

beta_init=0.09 #rate of infection
gamma=0.7 #recovery rate
days=90

inf=[]
sus=[]
rec=[]

def calc(beta, num_of_days):
    I=1
    R=0
    S=10000000 # the number of people without immunity
    alpha=I/(S+I+R)
    C=40 #the number of person to meet/contact

    for i in range(len(num_of_days)):
        #inf_tmp=[None for x in range(sum(num_of_days[0:i])-1)]
        #sus_tmp=[None for x in range(sum(num_of_days[0:i])-1)]
        #rec_tmp=[None for x in range(sum(num_of_days[0:i])-1)]
        inf_tmp=[]
        sus_tmp=[]
        rec_tmp=[]
        for t in range(num_of_days[i]):
            alpha=I/(S+I+R)#infection rate of contact person
            new_R=I*gamma
            new_I=S*C*alpha*beta[i]
            if new_I<0:
                new_I=0
            I=I+new_I-new_R
            R=R+new_R
            S=S-new_I
            if S<=0:
                S=0
            print(i,t,beta[i])
            inf_tmp.append(I)
            sus_tmp.append(S)
            rec_tmp.append(R)
        #print((inf_tmp))
        inf.append(inf_tmp)
        sus.append(sus_tmp)
        rec.append(rec_tmp)
      

calc([beta_init,beta_init*0.5], [10,days-10])

plt.subplot().set_xlim([0, days])
print(len(range(0, len(inf[0])+len(inf[1])-1)))
#plt.plot(range(0, len(inf[0])-1), inf[0],label='infection beta='+str(beta_init))
#plt.plot(range(len(inf[0]), len(inf[0])+len(inf[1])-1), inf[1],label='infection beta='+str(beta_init*0.7))
plt.plot(inf[0]+inf[1],label='infection')
plt.plot(rec[0]+rec[1],label='recover')
plt.plot(sus[0]+sus[1],label='susceptible')
#for item in inf: plt.plot(item,label='infection')
#for item in sus: plt.plot(item,label='susceptible')
#for item in rec: plt.plot(item,label='recover')

plt.legend()
```
